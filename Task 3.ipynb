## Predicting the Optimum number of clusters

```python
# getting the features of the data
x = data1.iloc[:, [0, 1, 2, 3]].values
# printing the feature
x
```

### K-Means

```python
# importing the required libraries
from sklearn.cluster import KMeans
```

```python
# creating a list within cluster sum of squares
lst=[]
j=range(1,11)
for i in j:
    means=KMeans(n_clusters= i)
    means.fit(x)
    lst.append(means.inertia_)
```

```python
# import the required libraries to plot
import matplotlib.pyplot as plt
%matplotlib inline
# plotting the values in a line graph with points
plt.plot(j,lst,'o-')
# plotted graph title
plt.title('The Elbow method showing the optimal number of clusters')
# x-axis label
plt.xlabel('Number of clusters')
# y-axis label
plt.ylabel('Cluster sum of squares')
```

```python
# Creating the kmeans classifier
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(x)
```

## Visualisation

```python
# Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 100, c = 'yellow', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 100, c = 'red', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 100, c = 'purple', label = 'Iris-virginica')
# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'black', label = 'Centroids')

plt.legend()
```
